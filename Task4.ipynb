```python
# Set up the subplots for numerical columns
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 15))

# Counter for subplot position
subplot_counter = 0

# Loop through numerical columns for univariate analysis
for i in range(4):
    for j in range(3):
        if subplot_counter < len(numerical_columns):
            col = numerical_columns[subplot_counter]
            
            # PDF and Histogram
            sns.histplot(data[col], kde=True, ax=axes[i, j])
            axes[i, j].set_title(f'{col} - PDF & Histogram')
            
            subplot_counter += 1

# Adjust layout
plt.tight_layout()
plt.show()

# Boxplots for numerical columns to identify outliers
plt.figure(figsize=(15, 10))
sns.boxplot(data=data[numerical_columns])
plt.title('Boxplots of Numerical Columns')
plt.show()

# Countplots for categorical columns
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))

# Counter for subplot position
subplot_counter = 0

# Loop through categorical columns for univariate analysis
for i in range(3):
    for j in range(3):
        if subplot_counter < len(categorical_columns):
            col = categorical_columns[subplot_counter]
            
            # Countplot
            sns.countplot(x=col, data=data, ax=axes[i, j])
            axes[i, j].set_title(f'{col} - countplot')
            axes[i, j].tick_params(axis='x', rotation=45,labelsize=8)
            
            subplot_counter += 1

# Adjust layout
plt.tight_layout()
plt.show()

# Identify outliers in each numerical column
for col in numerical_columns:
    # Calculate IQR
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Identify outliers
    outliers = ((data[col] < (Q1 - 1.5 * IQR)) | (data[col] > (Q3 + 1.5 * IQR)))
    
    # Print outliers for each numerical column
    print(f'Outliers in {col}: {data[outliers][col].tolist()}')
```
